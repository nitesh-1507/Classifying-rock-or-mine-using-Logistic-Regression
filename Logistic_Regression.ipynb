# Data Preprocessing and Logistic Regression - Sklearn

```python
# Package importing

import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import numpy as np
from AdvancedAnalytics import logreg
from sklearn.metrics import confusion_matrix
```

```python
# Data File importing

sonar=pd.read_excel('C:/Users/NITESH/Desktop/isen-656/sonar3by5.xlsx')
sonar.head()
```

```python
##Oultier detection

list=['R1', 'R2', 'R3', 'R4', 'R5', 'R6', 'R7', 'R8', 'R9', 'R10', 'R11',
       'R12', 'R13', 'R14', 'R15', 'R16', 'R17', 'R18', 'R19', 'R20', 'R21',
       'R22', 'R23', 'R24', 'R25', 'R26', 'R27', 'R28', 'R29', 'R30', 'R31',
       'R32', 'R33', 'R34', 'R35', 'R36', 'R37', 'R38', 'R39', 'R40', 'R41',
       'R42', 'R43', 'R44', 'R45', 'R46', 'R47', 'R48', 'R49', 'R50', 'R51',
       'R52', 'R53', 'R54', 'R55', 'R56', 'R57', 'R58', 'R59', 'R60']

def outlier(x):
    if x<0 or x>1:
        x=None
    else:
        x=x
    return x

for i in list:
    sonar[i]=sonar[i].apply(lambda x : outlier(x))
```

```python
##Imputing the missing value with the mean   

interval_data=sonar.as_matrix(columns=list)
interval_impute=preprocessing.Imputer(strategy='mean')
imputed_interval_data=interval_impute.fit_transform(interval_data)
sonar[list]=imputed_interval_data
sonar.head()
```

```python
##Train test split
X=np.asarray(sonar.select_dtypes(include=float))
Y=sonar.select_dtypes(include=object)
Y['object']=Y['object'].map({'R':1, 'M':0})
Y=np.asarray(Y)
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.3, random_state=101)
```

```python
##Logistic regression
log_fit=LogisticRegression()
log_fit.fit(X_train, Y_train)

# Displaying metrics on test data
logreg.display_binary_metrics(log_fit, X_test, Y_test)
```

```python
##Appending predicted class to original dataframe and displaying first 15 rows
sonar['Predicted_class']=log_fit.predict(X)
sonar['Predicted_class']=sonar['Predicted_class'].map({1:'R', 0:'M'})
sonar.head()
```
